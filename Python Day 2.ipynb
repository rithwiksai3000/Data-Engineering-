```python
# Reversing a string

s = 'Hritvik'
x = ''
for i in range(len(s)):
    x=s[i]+x
print(x)
```

```python
n=11
sum=0
i=1
while i<=n:
    sum=sum+i
    i+=1
print("Sum is " , sum)

# if counter value not used , the loop goes into infinite loop
```
