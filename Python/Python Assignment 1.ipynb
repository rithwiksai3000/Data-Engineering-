```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 1. Clean and Prepare the data
# We drop NA, split by comma, explode into individual rows, 
# then count and take the top 10 as a new DataFrame.
cuisine_counts = (
    df['Cuisines']
    .str.split(', ')
    .explode()
    .value_counts()
    .head(10)
    .reset_index()
)
cuisine_counts.columns = ['Cuisine', 'Count']

# 2. Visualize
plt.figure(figsize=(12, 6))

# Using barplot is safer here than countplot with an 'order' argument
sns.barplot(
    data=cuisine_counts,
    y='Cuisine',
    x='Count',
    hue='Cuisine',  # Assigning y to hue to satisfy new Seaborn rules
    palette='magma',
    legend=False
)

plt.title('Top 10 Most Frequent Cuisines in Zomato Dataset')
plt.xlabel('Number of Restaurants')
plt.ylabel('Cuisine Type')
plt.grid(axis='x', linestyle='--', alpha=0.7)
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# 1. Prepare the data with a "Unique Name"
# We sort by Rating and Votes to get a stable top 10
top_10 = df.sort_values(['Aggregate rating', 'Votes'], ascending=False).head(10).copy()

# Create a 'Display Name' combining Name and Locality so branches are distinct
top_10['Display Name'] = top_10['Restaurant Name'] + " (" + top_10['Locality'] + ")"

# 2. Set up the plot
plt.figure(figsize=(12, 8))

# Use the 'Display Name' for the Y-axis to ensure 10 unique bars
ax = sns.barplot(
    data=top_10,
    x='Aggregate rating',
    y='Display Name',
    hue='Display Name',
    palette='RdYlGn_r',
    legend=False
)

# 3. Fix the "floating" labels and alignment
plt.xlim(4.8, 5.0)  # Zoom in to see the differences

for p in ax.patches:
    width = p.get_width()
    ax.text(
        width + 0.002,         
        p.get_y() + p.get_height()/2, 
        f'{width:.1f}',        
        va='center', 
        fontweight='bold'
    )

plt.title('Top 10 Rated Restaurants (Individual Locations)', fontsize=15, pad=20)
plt.xlabel('Aggregate Rating', fontsize=12)
plt.ylabel('Restaurant (Locality)', fontsize=12)
plt.tight_layout()
plt.show()
```
