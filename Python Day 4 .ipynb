```python


import copy

original = [1,2,3,4] 

shallow = copy.copy(original)

shallow[0] = 12

print('original', original)
print('shallow', shallow)
```

```python


import copy

original = [[1,2],[3,4]] 

deep = copy.deepcopy(original)

deep[0][0] = 12

print('original', original)
print('Deep', deep)

```
