```python
def genfibon(n):
    a=1
    b=1
    for i in range(n):
        yield a # yielding something  , when we use yield , the function bevcomes a gnerator
        # yield remebers the postion valueof a 
        a,b= b,a+b
```

```python
genfibon(5) # initialize the object
```

```python
for num in genfibon(12):
    print(num)
```

```python
class Person:
    def __init__(self,name,surname,dob):
        self._name=name
        self._surname=surname
        self._dob=dob

    def age(self , current_year):
        return current_year - self._dob

    def __str__ (self):
        return "%s %s was born %d. "  % (self.__name , self.__surname , self.__dob)

alec=Person("Alec" , "Baldwin" , 1987)
# print(alec._Person__dob) # now we ca access the method but not how the method is calculating
print(alec.age(2026))
# print(alec.__surname)
```

```python
class Student(Person):
    def __init__(self,student_id , *args):
        super(Student , self). __init__(*args)

        self._student_id = student_id
charlie = Student(1,'Charlie', 'Brown' , 2006)

print(charlie._student_id)
print(type(charlie))
print(charlie._first_name)
print(charlie._last_name)
print(ininstance(charlie , Person))
print(isinstance(charlie , object))
```
